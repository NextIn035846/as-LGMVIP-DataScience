```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
img = cv2.imread('avengers-endgame-imax-poster-crop.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis('off')
plt.title('Original image')
plt.show()
```

```python
ima_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(ima_gray,cmap='gray')
plt.axis('off')
plt.title('GrayScale Image')
plt.show()
```

```python
img_invert = cv2.bitwise_not(ima_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap='gray')
plt.axis('off')
plt.title('Inverted Image')
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert,(21,21),sigmaX =0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
final = cv2.divide(ima_gray,255-img_smoothing,scale=256.0)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
